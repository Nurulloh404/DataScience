# Exercise 00: Load and Save

```python
import pandas as pd
```

```python
# Load the dataset
# The file is located in ../data/feed-views.log relative to this notebook
df = pd.read_csv('../data/feed-views.log', 
                 skiprows=[2, 3], 
                 skipfooter=2, 
                 names=['datetime', 'user'], 
                 engine='python') 
# engine='python' is needed for skipfooter
```

```python
# Set datetime as index
df = df.set_index('datetime')
```

```python
# Rename datetime to date_time (index name)
df.index.name = 'date_time'
```

```python
# Save to csv with semicolon delimiter
df.to_csv('feed-views-semicolon.log', sep=';')
```
